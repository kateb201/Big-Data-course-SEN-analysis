## Imports

```python
import tweepy
from confluent_kafka import Producer  
```

## Producer Configuration

```python
p = Producer({'bootstrap.servers': 'localhost:8080'}) 
consumer_key = "KEY_123"
consumer_secret = "CONS_SEC"
access_token = "ACC_TOK"
access_token_secret = "ACC_TOK_SEC"

bearer_token = "BEAR_TOK"

client = tweepy.Client(bearer_token=bearer_token)
query = 'news'
response = client.search_recent_tweets(query=query,max_results=100,tweet_fields=['lang'])

counter=1
tweetLimit=100
tweetList = ['']
flag=False

for tweet in tweepy.Paginator(client.search_recent_tweets,query=query,max_results=100,tweet_fields=['lang']).flatten(limit=tweetLimit):
    if (tweet.lang == 'en'):
        for i in tweetList:
            if i == tweet.text:
                flag=False
                break
        if flag:
            p.produce('twitter',key='myKey',value=tweet.text) #produce the msg to kafka
            print(counter," : ",tweet.text)
            tweetList.append(tweet.text)
            counter = counter + 1
        else: tweetLimit=tweetLimit+1
        flag=True    
```
